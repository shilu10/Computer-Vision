# **Required Libraries**

```python
import easyocr
import cv2 
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read the images 

img_path = '/content/one_text.png'

# Create a easyocr reader 
easy_reader = easyocr.Reader(['en'], gpu = True)

# With the reader, we can read the text 
text = easy_reader.readtext(img_path)

# result
text
```

```python
img = cv2.imread(img_path)
plt.imshow(img)
```

```python
top_left = text[0][0][0]
bottom_right = text[0][0][2]
text_in = text[0][1]
font = cv2.FONT_HERSHEY_SIMPLEX
```

```python
img = cv2.rectangle(img, top_left, bottom_right, (0, 0, 188), 2)
img = cv2.putText(img, text_in, top_left , font, .5, (255, 15, 25), 2)
```

```python
plt.figure(figsize = (10, 10))
plt.imshow(img)
```

```python
img_path = '/content/traffic.jpeg'

reader = easyocr.Reader(['en'], gpu = True)
result = reader.readtext(img_path)
result
```

```python
img = cv2.imread(img_path)
for res in result : 
  top_left = res[0][0]
  bottom_right = res[0][2]
  text = res[1]
  font = cv2.FONT_HERSHEY_SIMPLEX
  img =  cv2.rectangle(img, top_left, bottom_right, (0, 255, 0), 2)
  # attributes --> image, text you need to put, where you needed to put, font, size of the text, color of the text, thickness of the text
  img = cv2.putText(img, text, top_left, font, 0.4, (255, 0, 5), 1)

plt.figure(figsize = (13, 13))
plt.imshow(img)
plt.show()
```
