```python
import keras_ocr as k_ocr
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
# this is used to download the weights of the keras_ocr pretrained model
k_pipeline = k_ocr.pipeline.Pipeline()
```

```python
# We are reading the image in the format of keras_ocr
images = [k_ocr.tools.read(image) for image in [
              '/content/traffic.jpeg',
              '/content/one_text.png'
]]
```

```python
# We gonna pass these embedded images to the recoginzer mthod in keras_ocr which is responsible for converting the images into text
```

```python
predictions = k_pipeline.recognize(images)
```

```python
predictions
```

```python
# picture
import cv2 
img = cv2.imread('/content/traffic.jpeg')

plt.imshow(img)
```

```python
fig, axs = plt.subplots(nrows=len(images), figsize=(10, 10))
for ax, image, prediction in zip(axs, images, predictions):
    k_ocr.tools.drawAnnotations(image=image, predictions=prediction, ax=ax)
```
